```python
#using colormap 
x=np.array([2,3,4,5,6,7,6,9,8,8,7,10])
y=np.array([4,5,6,7,8,9,7,6,8,9,0,5])
colors=np.array([10,30,90,80,60,95,48,89,78,80,10,5])
size=np.array([10,90,80,70,50,40,80,79,67,80,70,500])
plt.scatter(x,y,c=colors,cmap="viridis",s=size)
plt.colorbar()
plt.show()
```

```python
x=np.random.randint(50,size=(50))
y=np.random.randint(50,size=(50))
colors=np.random.randint(50,size=(50))
sizes=20*np.random.randint(50,size=(50))
plt.scatter(x,y,c=colors,s=sizes,cmap="nipy_spectral",alpha=0.5)
plt.colorbar()

plt.show()
```

```python
#bar graph
x=np.array(["manasa","swaroop","praveen","ruchi","ramya"])
y=np.array([90,100,80,90,70])
colors=["r","m","k","y","c"]
plt.bar(x,y,width=0.5,color=colors)
plt.xlabel("NAMES",color="m")
plt.ylabel("MARKS",color="m")
plt.title("STUDNET MARKS",color="r")
plt.grid(color="green",linestyle="-.")
plt.show()
```

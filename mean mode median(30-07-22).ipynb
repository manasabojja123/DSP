```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
p1=pd.Series([1,2,3,4,8,8,7])
p2=pd.Series([2,4,5,6,6,8,9])
p3=pd.Series([5,6,7,8,8])
p4=pd.Series([50,80,90,70,60,80])
p5=pd.Series([50,60,60,80,90])
print("_____p1_____")
print(p1.mean())
print(p1.mode())
print(p1.median())
print(p1.std())
print(p1.var())
print("_____p2_____")
print(p2.mean())
print(p2.mode())
print(p2.median())
print(p2.std())
print(p2.var())
print("_____p3_____")
print(p3.mean())
print(p3.mode())
print(p3.median())
print(p3.std())
print(p3.var())
print("_____p4_____")
print(p4.mean())
print(p4.mode())
print(p4.median())
print(p4.std())
print(p4.var())
print("_____p5____")
print(p5.mean())
print(p5.mode())
print(p5.median())
print(p5.std())
print(p5.var())



```

```python
x=pd.DataFrame({
    "MEAN":[p1.mean(),p2.mean(),p3.mean(),p4.mean(),p5.mean()],
    "MODE":[p1.mode(),p2.mode(),p3.mode(),p4.mode(),p5.mode()],
    "MEDIAN":[p1.median(),p2.median(),p3.median(),p4.median(),p5.median()]
},index=["P1","P2","P3","P4","P5"])
x
```

```python
#MEAN pie
plt.pie(x["MEAN"],labels=["p1","p2","p3","p4","p5"])
plt.legend(title="MEAN")
plt.show()
#MEAN HIST
plt.hist(x["MEAN"],bins=200,label="MEAN",color="hotpink")
plt.show()
#MEAN BAR
plt.bar(x["MEAN"],[1,2,3,4,5],color=["green","pink","yellow","red","black"],width=5,label=["p1","p2","p3","p4","p5"])
plt.legend(title="MEAN")
plt.title("MEAN",color="red",size=30)
plt.show()

```

```python
#Mode pie
plt.pie(x["MODE"],labels=["p1","p2","p3","p4","p5"])
plt.legend(title="MODE")
plt.show()
#MODE HIST
plt.hist(x["MODE"],bins=500,label="MEAN")
plt.show()
#MODE BAR
plt.bar(x["MODE"],[1,2,3,4,5],color=["green","pink","yellow","red","black"],width=5,label=["p1","p2","p3","p4","p5"])
plt.legend(title="MODE")
plt.title("MODE",color="red",size=30)
plt.show()

```

```python
#MEAN pie
plt.pie(x["MEDIAN"],labels=["p1","p2","p3","p4","p5"])
plt.legend(title="MEDIAN")
plt.show()
#MEAN HIST
plt.hist(x["MEDIAN"],bins=200,label="MEDIAN",color="hotpink")
plt.show()
#MEAN BAR
plt.bar(x["MEDIAN"],[1,2,3,4,5],color=["green","pink","yellow","red","black"],width=5,label=["p1","p2","p3","p4","p5"])
plt.legend(title="MEDIAN")
plt.title("MEDIAN",color="red",size=30)
plt.show()

```
